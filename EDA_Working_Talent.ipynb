```python
#missing data
total = google_playstore.isnull().sum().sort_values(ascending=False)
percent = (google_playstore.isnull().sum()/google_playstore.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
# missing data

google_playstore = google_playstore.drop((missing_data[missing_data['Total'] > 1]).index,1)
google_playstore = google_playstore.drop(google_playstore.loc[google_playstore['Rating'].isnull()].index)
```
